# Case Study: Data‑Driven Insights in the U.S. Office Market (2025)

**1. Key News Story & Context**

In 2025, data from CBRE reveals a major structural shift in the U.S. office market: office conversions and demolitions will exceed new construction, signaling a potential recovery mechanism by removing obsolete or underused supply. *(CBRE Research, June 2025)*


CBRE’s analysis across 58 major U.S. markets projects 23.3 million square feet (MSF)of office space will be converted or demolished in 2025. In contrast, only 12.7 MSF of new office space is expected to be completed in those same markets this year. The pipeline for future conversion is massive, with 81 MSF of office space planned or underway for conversion in the coming years.


The national office vacancy rate remains high, around 19%, according to CBRE’s report. Positive net absorption has been recorded in recent quarters: CBRE notes four straight quarters of positive net absorption after a long decline.


However, the CBRE report also flags significant headwinds such as rising construction costs, labor constraints, and financing challenges could limit conversion momentum. This represents a data‑driven turning point in the U.S. office real estate market — not just a decline, but a fundamental adjustment in supply.

**2. The Role of Data Analysis**

In this case, data analysis is central to how market participants understand and act on the transformation:

**Vacancy & Supply Monitoring**

- CBRE uses its proprietary CRE‑market data across 58 large U.S. markets to calculate vacancy, supply coming offline, and supply being added.

- This allows real-time (or near-term) tracking of how much office space is being removed versus added, which is a critical leading indicator for future pricing and demand.

**Conversion & Demolition Feasibility Analysis**

- The data includes a pipeline of 81 MSF for conversion, used to model long-term removal and reuse of assets.

- Analysts can segment this pipeline by market (e.g., Manhattan, DC, Houston), which helps prioritize which buildings or regions are most likely to be converted or demolished. According to CBRE, Manhattan tops the list with 10.3 MSF earmarked for conversion.

- They also break down the type of conversions: for example, a large share is expected to become multifamily, though conversion economics, zoning, and cap structure may differ.

**Financial Modeling & Risk Assessment**

- With conversion pipeline data + expected removal + new construction forecasts, investors and developers can run underwriting models that account for cost of demolition or conversion, expected returns, cash flow from repurposed assets, and risk scenarios (e.g., regulatory changes, cost overruns).

- The report underscores that financing remains a risk as high construction costs and labor constraints are nontrivial. Lenders can use these models to decide which CRE loans to underwrite, not just on stabilized office assets, but on conversion plays.

**Trend Forecasting & Strategic Planning**

- By combining vacancy trends (19%), pipeline data (81MSF), and removal projections (23.3MSF), CBRE and other market actors can forecast how the U.S. office market might rebalance.
- These forecasts feed into long-term strategy like which markets to exit, which assets to convert, which to hold, which to demolish.
- The data also helps predict how supply dynamics may impact future vacancy and rental growth, particularly for high-quality (“prime”) properties.

**3. Impact of Data Analysis**

**Strategic Decision-Making**

- Developers and investors are less likely to rely on speculative bets; they can target specific properties or submarkets for conversion, backed by hard data.

- Capital can be allocated more efficiently: funds may go to properties that are viable for re-use (or demolition) rather than chasing new construction in an oversupplied market.

**Risk Management**

- Lenders, armed with data, can underwrite with more nuance: properties with high conversion potential may present lower long-term risk than “zombie” offices likely to remain vacant.

- Investors can stress-test conversion returns under different cost and demand scenarios, reducing downside risk.

**Market Rebalancing**

- The net removal of 23 MSF helps reduce supply, which could gradually ease vacancy pressure. This may contribute to a more sustainable market recovery.

- The shrinking pipeline could benefit surviving, high-quality office buildings by reducing competition from obsolete inventory.

Without data analysis:

- Stakeholders might misjudge which assets to convert, leading to capital misallocation.

- Lenders might be overly conservative or overly aggressive, increasing risk.

- The market could remain stuck in oversupply, delaying recovery or causing malinvestment.

**4. Conclusion**

This 2025 CBRE‑based case illustrates how deep, data-driven analysis is not just supporting decision-making — it's shaping the very direction of the U.S. office market. Rather than building more “new” offices, stakeholders are using data to remove or repurpose space, stabilize supply, and manage long-term risk.
