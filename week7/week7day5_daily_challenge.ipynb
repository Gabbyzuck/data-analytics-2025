```python
import numpy as np
from PIL import Image
from scipy.ndimage import rotate
import matplotlib.pyplot as plt

image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

image_array = np.array(original_image)

def rotate_image_30_degrees(image_array):
    # rotate(image, angle, reshape=True/False (to keep dimensions), mode='constant'/'nearest' etc.)
    return rotate(image_array, 30, reshape=True)
```

```python
# Call the function with image array
rotated_array = rotate_image_30_degrees(image_array)

# Convert rotated array back to PIL image for display
rotated_image = Image.fromarray(rotated_array.astype(np.uint8))

# Show the rotated image using matplotlib
plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Flip horizontally (mirror)
horizontal = ImageOps.mirror(original_image)

# Flip vertically (flip)
vertical = ImageOps.flip(original_image)

# Show all images side by side
fig, axes = plt.subplots(1, 3, figsize=(15,5))

axes[0].imshow(original_image)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(horizontal)
axes[1].set_title('Horizontal Image')
axes[1].axis('off')

axes[2].imshow(vertical)
axes[2].set_title('Vertical Image')
axes[2].axis('off')

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Get original size
width, height = original_image.size

# Calculate new size (1.2 times bigger)
new_size = (int(width * 1.2), int(height * 1.2))

# Resize (zoom in)
zoomed_image = original_image.resize(new_size, resample=Image.LANCZOS) #LANCZOS is high quality resize

# Display using matplotlib
plt.imshow(zoomed_image)
plt.axis('off')  # hide axes for better view
plt.show()
```
