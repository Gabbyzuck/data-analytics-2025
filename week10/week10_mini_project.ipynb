```python
import pandas as pd
df = pd.read_csv('Palmer Penguins Dataset.csv')
df_cleaned = df.dropna()
```

```python
import matplotlib.pyplot as plt
import numpy as np

species_count = df_cleaned.groupby(['species', 'sex']).size().unstack()

x = np.arange(len(species_count.index))
width = 0.35

plt.bar(x - width/2, species_count['male'], width, label='male')
plt.bar(x + width/2, species_count['female'], width, label='female')

plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Penguin Count by Sex')
plt.xticks(x, species_count.index)
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

species_matric = species_count.corr()

sns.heatmap(species_matric, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('Palmer Penguins Dataset.csv')
```

```python
df_cleaned = df.dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df_cleaned['bill_length_mm'])
plt.title('Histogram of Bill Length')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(df_cleaned, col = 'species', row = 'sex')
g.map(sns.boxplot, 'flipper_length_mm', 'body_mass_g')
plt.show()
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,8))
sns.histplot(df_cleaned['bill_length_mm'], ax=ax[0,0])
sns.boxplot(df_cleaned['bill_depth_mm'], ax=ax[0,1])
```

```python
sns.pairplot(df_cleaned[['flipper_length_mm', 'bill_length_mm', 'body_mass_g']])

g = sns.PairGrid(df_cleaned[['flipper_length_mm', 'bill_length_mm', 'body_mass_g']])
g.map_upper(sns.scatterplot)
g.map_lower(sns.boxplot)
g.map_diag(sns.scatterplot)
```
