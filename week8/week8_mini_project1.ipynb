```python
import matplotlib.pyplot as plt
import numpy as np

cleaned_data = health_data[(health_data['Age'] >= 18) & (health_data['Age'] <= 100)]

plt.hist(cleaned_data['Age'], bins=10, color='blue')
plt.title('Mental Health Conditions Among Various Ages')
plt.xlabel('Age')
plt.ylabel('Conditions')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = health_data['Gender']

plt.hist(health_data['Gender'], bins=10, color='red')
plt.title('Mental Health Conditions Among Various Genders')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

recorded_country = health_data['Country'].value_counts()

plt.figure(figsize=(12,6))
plt.bar(recorded_country.index, recorded_country.values)
plt.title('Health Issues Reported By Country')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.show()
```
