```python
#Exercise 3: Feature Engineering
import pandas as pd

titanic_data = pd.read_csv('titanic dataset/train.csv')

titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1  # +1 for the passenger
titanic_data['Title'] = titanic_data['Name'].str.extract(r',\s*([^\.]+)\.')

print(titanic_data[['Name', 'Title', 'SibSp', 'Parch', 'FamilySize']].head())
```

```python
#Exercise 4: Outlier Detection And Handling
import pandas as pd

titanic_data = pd.read_csv('titanic dataset/train.csv')

Q1 = titanic_data['Fare'].quantile(0.25)
Q3 = titanic_data['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

titanic_data = titanic_data[(titanic_data['Fare'] >= lower_bound) & (titanic_data['Fare'] <= upper_bound)]

print(titanic_data['Fare'].describe())
```

```python
import pandas as pd
import numpy as np

titanic_data = pd.read_csv('titanic dataset/train.csv')

Q1 = titanic_data['Age'].quantile(0.25)
Q3 = titanic_data['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping outliers
titanic_data['Age'] = titanic_data['Age'].clip(lower=lower_bound, upper=upper_bound)

print(titanic_data['Age'].describe())
```

```python
from scipy.stats import zscore
import pandas as pd

titanic_data = pd.read_csv('titanic dataset/train.csv')

# Drops missing Age values first
titanic_data = titanic_data.dropna(subset=['Age'])

# Calculates Z-scores
titanic_data['z_age'] = zscore(titanic_data['Age'])

# Keeps only data within 3 standard deviations
titanic_data = titanic_data[(titanic_data['z_age'].abs() < 3)]

print(titanic_data['Age'].describe())
```

```python
#Exercise 7: Data Transformation For Age Feature
import pandas as pd

titanic_data = pd.read_csv('titanic dataset/train.csv')

bins = [0, 10, 20, 30, 40, 50]
labels = ['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']

# Creates a new column 'AgeGroup'
titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'], bins=bins, labels=labels, right=False)

print(titanic_data[['Age', 'AgeGroup']].head(15))
```

```python
agegroup_dummies = pd.get_dummies(titanic_data['AgeGroup'], prefix='AgeGroup')

titanic_data = pd.concat([titanic_data, agegroup_dummies], axis=1)

print(titanic_data.head())
```
