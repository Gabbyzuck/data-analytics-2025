```python
import pandas as pd

df = pd.read_csv('2019.csv')

df = df[['Country or region', 'GDP per capita', 'Healthy life expectancy']]




```

```python
import matplotlib.pyplot as plt

fig, (ax1) = plt.subplots(figsize=(14, 6))

ax1.bar(df['Country or region'], df['GDP per capita'])
ax1.set_ylabel('GDP per capita')
ax1.set_xticklabels(df['Country or region'])

ax1.plot(df['Country or region'], df['Healthy life expectancy'], color='green', marker='o', label='Healthy life expectancy')

plt.title('GDP per Capita & Healthy Life Expectancy by Region')
plt.show()
```
