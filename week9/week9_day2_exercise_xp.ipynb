```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Student Mental health.csv')

sns.histplot(df['What is your CGPA?'], color='red')
plt.title('CGPA')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.barplot(x='Choose your gender', y='Do you have Anxiety?', data=df)
plt.title('Anxiety by Gender')
plt.show()
```

```python
#Exercise 3: Exploring the Relationship Between Age and Panic Attacks

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df['Do you have Panic attack?'] = df['Do you have Panic attack?'].replace({'Yes': 1, 'No': 0})

sns.lmplot(x='Age', y='Do you have Panic attack?', data=df)
plt.title('Panic Attacks by Age')
plt.xlabel('Age')
plt.xlabel('Number of Panic Attack Responses')
plt.show()
```

```python
#Exercise 4: Visualizing Pairwise Relationships with Seaborn Pair Plot

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cols = ['Age', 'What is your CGPA?', 'Do you have Anxiety?',
        'Do you have Depression?', 'Do you have Panic attack?']

df[['Do you have Anxiety?', 'Do you have Depression?', 'Do you have Panic attack?']] = df[[
    'Do you have Anxiety?', 'Do you have Depression?', 'Do you have Panic attack?'
]].replace({'Yes': 1, 'No': 0})


sns.pairplot(df[cols], hue='Age', markers=['o', 'x', 'p', 's'])
plt.show()
```

```python
#Exercise 5: Creating a Heatmap to Visualize Correlations

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

correlation_materix = df[cols].corr()

sns.heatmap(correlation_materix, annot=True, cmap='coolwarm')
plt.show()

```

```python
#Exercise 6: Analyzing Distributions and Relationships Using FacetGrid

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df['What is your CGPA?'] = pd.to_numeric(df['What is your CGPA?'])

g = sns.FacetGrid(df, col='Do you have Depression?', height=4)
g.map(sns.histplot, 'What is your CGPA?', bins=10, color='skyblue')

g.set_axis_labels('CGPA', 'Count')
g.set_titles(col_template='Depression: {col_name}')
plt.show()
```
